```python
import IPython.display
import io
```

```python
def show_cairo_canvas(cairo_canvas):
    with io.BytesIO() as fileobj:
        cairo_canvas.write_to_png(fileobj)
        IPython.display.display(IPython.display.Image(fileobj.getvalue()))
```

```python
cairo_canvas = cairo.ImageSurface(cairo.FORMAT_ARGB32, 640, 480)
cairo_ctx = cairo.Context(cairo_canvas)
```

```python
show_cairo_canvas(cairo_canvas)
```

```python
cairo_ctx.set_source_rgba(1.0, 1.0, 1.0, 1.0)
cairo_ctx.set_line_cap(cairo.LINE_CAP_ROUND)
cairo_ctx.set_line_width(20)

cairo_ctx.move_to(0, 0)
cairo_ctx.line_to(300, 300)
cairo_ctx.stroke()
show_cairo_canvas(cairo_canvas)
```

```python
cairo_ctx.set_source_rgba(1.0, 1.0, 0, 1.0)
cairo_ctx.set_line_cap(cairo.LINE_CAP_ROUND)
cairo_ctx.set_line_width(20)

cairo_ctx.move_to(500, 0)
cairo_ctx.line_to(0, 480)
cairo_ctx.stroke()
show_cairo_canvas(cairo_canvas)
```

```python
cairo_ctx.set_source_rgba(0, 1.0, 0, 1.0)
cairo_ctx.set_line_cap(cairo.LINE_CAP_ROUND)
cairo_ctx.set_line_width(20)

cairo_ctx.move_to(120, 120)
cairo_ctx.line_to(500, 400)
cairo_ctx.stroke()
show_cairo_canvas(cairo_canvas)
```
